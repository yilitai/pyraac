```python
from pathlib import Path
import json

BASE_DIR = Path("./").resolve()
RT_DESC_PATH = BASE_DIR / Path("./pyraac/json") / "rt_desc.json"
RT_OPTION_PATH = BASE_DIR / Path("./pyraac/json") / "rt_option.json"

with open(RT_DESC_PATH, "r", encoding="utf-8") as file:
    temp_rt_desc_dict = json.load(file)

with open(RT_OPTION_PATH, "r", encoding="utf-8") as file:
    temp_rt_option_dict = json.load(file)
    
temp_rt_option_dict
```

```python
max_raa_value = 0

for item in temp_rt_option_dict:
    # print(item['raaClusters'])
    for child in item['raaClusters']:
        # print(child)
        if child['raaValue'] > max_raa_value:
            max_raa_value = child['raaValue']
            
max_raa_value
```

```python
rt_toparent = {}
for item in temp_rt_option_dict: 
    children_values = [cluster['raaValue'] for cluster in item['raaClusters']]
    # print(children_values)
    # print(item['raaValue'])
    for child in children_values:
        rt_toparent[child] = item['raaValue']
        
# rt_toparent

with open(BASE_DIR / Path("./pyraac/json") / "rt_toparent.json", 'w', encoding='utf-8') as file:
    json.dump(rt_toparent, file, ensure_ascii=False, indent=4)
```
